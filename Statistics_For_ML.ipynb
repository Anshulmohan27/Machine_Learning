Mean = Average value of the given data. Sum all the values and divide by the number of values.

Median = Middle value of an ordered data set. If even number of data, compute mean of middle 2 values.

Interview Question

"When would you prefer median over mean?"

Good answer:

When data contains significant outliers or is heavily skewed.

Variance = How spread out the data is. Take the mean, and then the difference with each value, square it, and add them up, and finally divide by the number of values.

Why Squaring?

Three reasons:

1. Removes negatives
(-4)^2 = 16

No cancellation.

2. Penalizes large deviations
1 away → 1
3 away → 9
10 away → 100

Large errors matter much more.

3. Leads to nice mathematics

Many ML algorithms rely on squared error because it makes optimization easier.

Standard Deviation = Square root of variance. Measures spread but in same dimensions as the data.

Z-Score =  (x-mean)/std deviation

What does a z-score of 3 mean?

Expected answer:

The observation is three standard deviations above the mean.
Highly unusual.

Important ML Insight

Suppose you have a dataset:

Feature	Value
Age	40
Salary	₹500,000

Raw numbers suggest salary is much larger.

After converting to z-scores:

Feature	Z-score
Age	1.2
Salary	0.8

Now both are measured on the same scale.

This is why standardization is one of the first preprocessing steps in ML.

Probablity = Meausure of the likelihood of an event.

Independent Events

This is one of the most important ideas in probability.

What does "independent" mean?

Two events are independent if:

Knowing the outcome of one event tells you nothing about the other.

The Multiplication Rule

For independent events: P(A∩B)=P(A)⋅P(B)

Naive Bayes

It's called "naive" because it assumes features are independent.

They usually aren't, but the method still works surprisingly well.

Dependent Events

Not everything is independent.

Suppose a bag contains:

3 red
2 blue

You draw a ball without replacement.

First draw:

P(red) = 3/5

Suppose you got a red.

Now the bag contains:

2 red
2 blue

Second draw:

P(red) = 2/4 = 1/2

The second probability changed.

Therefore the events are dependent. So probability is (3/5 + 1/2)

Conditional Probability = Probability of a given event given some condition is true and defined.

Bayes' Theorem

This is probably the single most famous equation in probability.

The problem Bayes solves is:

We know how likely evidence is if something is true.

But what we really want is the probability that something is true given the evidence.

P(A∣B)=P(B∣A).P(A)​/P(B)

What does Bayes' theorem do?

Good answer:

Bayes' theorem updates the probability of a hypothesis when new evidence becomes available.

Central Limit Theorem (CLT)

One of the most important ideas in statistics.

It says:

If you repeatedly sample data and calculate averages, those averages become normally distributed.

Even if original data isn't normal.

This is why normal distributions appear everywhere.
